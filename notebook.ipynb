The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

Question 1: What is the most commonly awarded gender and birth country?
Store your answers as string variables top_gender and top_country.

```python
top_gender = nobel['sex'].mode().iloc[0]
top_gender
```

```python
top_country = nobel['birth_country'].mode().iloc[0]
top_country
```

Question 2: Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?
Store this as an integer called max_decade_usa.

```python
nobel['us_winners'] = nobel['birth_country'].isin(['United States of America', 'USA'])
nobel.head()
```

```python
nobel['decade'] = (np.floor(nobel['year'] / 10) * 10).astype(int)
prop_usa_winners = nobel.groupby('decade', as_index=False)['us_winners'].mean()
```

```python
prop_usa_winners
```

```python
max_decade_usa = prop_usa_winners[prop_usa_winners['us_winners'] == prop_usa_winners['us_winners'].max()]['decade'].values[0]
max_decade_usa
```

Question 3: Which decade and Nobel Prize category combination had the highest proportion of female laureates?

Store this as a dictionary called max_female_dict where the decade is the key and the category is the value. There should only be one key:value pair.

```python
nobel['female_winner'] = nobel['sex'] == 'Female'
prop_female_winners = nobel.groupby(['decade', 'category'], as_index=False)['female_winner'].mean()
prop_female_winners
```

```python
female_decade_cat = prop_female_winners[prop_female_winners['female_winner']== prop_female_winners['female_winner'].max()][['decade', 'category']]
```

```python
max_female_dict = {female_decade_cat['decade'].values[0]: female_decade_cat['category'].values[0]}
max_female_dict
```

Question 4: Who was the first woman to receive a Nobel Prize, and in what category?
Save your string answers as first_woman_name and first_woman_category.

```python
female_laureates = nobel[nobel['sex'] == 'Female']

first_woman = female_laureates.sort_values(by='year').iloc[0]
first_woman_name = first_woman['full_name']
first_woman_category = first_woman['category']
```

```python
first_woman_name, first_woman_category
```

```python
count_nobel = nobel['full_name'].value_counts()
repeat_nobel = count_nobel[count_nobel>=2].index
repeat_list = list(repeat_nobel)
repeat_list
```
